## News Content extraction

```python
from newspaper import Article, ArticleException
import csv
from pathlib import Path
```

```python
STOCK = "AMZN"
STOCK_NAME = "Amazon"
```

```python
news_csv = Path(f"../Datasets/Twitter URL/{STOCK}-filtered.csv")
```

```python
done = set()
news_contents = []

with open(news_csv, 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        url, time = row['url'], row['time']
        if url not in done:
            print(f"URL - {url}")
            done.add(url)
            try:
                article = Article(url)
                article.download()
                article.parse()
                content = article.text.replace("\n", " ").lower()
            except ArticleException:
                print(f"Error on {url}")
                continue
                
            if STOCK_NAME.lower() not in content:
                continue
            
            news_contents.append({
                "content": content,
                "time": time
            })
            print("Success\n")
```

```python
len(news_contents)
```

```python
target = Path(f"../Datasets/Extracted News Contents/{STOCK}.csv")

with open(target, 'w') as f:
    fieldnames = ['time', 'content']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    
    for item in news_contents:
        writer.writerow(item)
```
